```python
class node:
  def __init__(self, name, age, p_id):
    self.name=name
    self.age=age
    self.p_id=p_id
    self.next=None

class PatientList:
  def __init__(self):
    self.head=None

  def insert_pat(self, name, age, p_id):
    new_node=node(name, age, p_id)
    if not self.head:
      self.head=new_node
    else:
      current = self.head
      while current.next:
        current = current.next
      current.next=new_node
    print(f"Inserted Name: {name}, Age:{age}, Pateint:{p_id}")

  def delete_pat(self, p_id):
    current=self.head
    prev=None
    while current:
      if current.p_id==p_id:
        if prev:
          prev.next=current.next
        else:
          self.head=current.next
        print(f"\nDelete Patient:{p_id}")
        return
      prev = current
      current = current.next
    print(f"Pateint with {p_id} not found")

  def display_pat(self):
    current=self.head
    if not current:
      print("No Patients")
      return
    print("\nPatients list:")
    while current:
      print(f"Name:{current.name}, Age:{current.age}, P_ID:{current.p_id}")
      current=current.next

plist=PatientList()
plist.insert_pat("Ramu",55,100)
plist.insert_pat("bhanu",30,101)
plist.insert_pat("Somu",35,102)
plist.display_pat()
plist.delete_pat(101)
plist.display_pat()
```
